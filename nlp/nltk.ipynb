# NLTK
A single notebook covering core NLTK operations and a small corpus-analysis workflow on Shakespeare text.

## Setup
Install NLTK if needed and download the resources used in this notebook.

## Basic Text Processing

```python
example_sentence = "Dr. Smith went to Washington. He arrived late."
word_tokenize(example_sentence), sent_tokenize(example_sentence)
```

```python
sample_text = "Hello! Welcome to the world of Natural Language Processing. Let's learn NLTK together."
stop_words = set(stopwords.words("english"))
filtered_tokens = [token for token in word_tokenize(sample_text) if token.lower() not in stop_words]
filtered_tokens
```

```python
word_samples = ["running", "jumps", "easily", "programming", "programmers"]
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

stemmed_words = [stemmer.stem(word) for word in word_samples]
lemmatized_words = [lemmatizer.lemmatize(word) for word in word_samples]

stemmed_words, lemmatized_words
```

```python
tagged_tokens = pos_tag(word_tokenize(sample_text))
tagged_tokens
```

```python
ne_chunk(tagged_tokens)
```

## Corpus Analysis
These cells expect a local `pg100.txt` file containing the Shakespeare corpus.

```python
corpus_path = Path("pg100.txt")
text = corpus_path.read_text(encoding="utf-8")

tokens = [token for token in word_tokenize(text) if token.isalpha()]
sentences = sent_tokenize(text)

print(f"Word count: {len(tokens)}")
print(f"Sentence count: {len(sentences)}")
```

```python
word_lengths = [len(word) for word in tokens]
print({
    "min": min(word_lengths),
    "max": max(word_lengths),
    "avg": round(sum(word_lengths) / len(word_lengths), 2),
    "median": median(word_lengths),
})
```

```python
stemmed_vocabulary = {stemmer.stem(word.lower()) for word in tokens}
print(f"Unique stemmed words: {len(stemmed_vocabulary)}")
```

```python
sentence_lengths = [len(word_tokenize(sentence)) for sentence in sentences]
shortest_sentence = min(sentences, key=lambda sentence: len(word_tokenize(sentence)))
longest_sentence = max(sentences, key=lambda sentence: len(word_tokenize(sentence)))

print({
    "min": min(sentence_lengths),
    "max": max(sentence_lengths),
    "avg": round(sum(sentence_lengths) / len(sentence_lengths), 2),
    "median": median(sentence_lengths),
})
print("Shortest sentence:", shortest_sentence)
print("Longest sentence:", longest_sentence)
```

```python
nouns = [word.lower() for word, pos in pos_tag(tokens) if pos.startswith("NN")]
FreqDist(nouns).most_common(10)
```
